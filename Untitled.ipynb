```python
from statsmodels.tsa.stattools import adfuller
```

```python
test_fuller=adfuller(df["Sales"])
```

```python
def adfuller_test(sales):
    result=adfuller(sales)
    labels=['ADF Test Statistic','P value','#Tags used','Number of observation']
    for value,label in zip(result,labels):
        print(label+':'+str(value))
    if result[1]<= 0.05:
        print("Stronge evidence and stationary")
    else:
        print("Not stronge evidence and Non stationary")
```

```python
adfuller_test(df['Sales'])
```

```python
df['Difference in sales']=df['Sales']-df['Sales'].shift(1)
```

```python
adfuller_test(df['Difference in sales'].dropna())
```

```python
df['Seasonal_sale']=df['Sales']-df['Sales'].shift(12)
```

```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
import statsmodels.api as sm
```

```python
fig=plt.figure(figsize=(12,8))
ax1=fig.add_subplot(211)
fig=sm.graphics.tsa.plot_acf(df['Seasonal_sale'].iloc[13:],lags=40,ax=ax1)
ax2=fig.add_subplot(212)
fig=sm.graphics.tsa.plot_pacf(df['Seasonal_sale'].iloc[13:],lags=40,ax=ax2)
```

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
model = sm.tsa.arima.ARIMA(df['Sales'], order=(1,1,1))
result = model.fit()
```

```python
result.summary()
```

```python
df
```

```python
df['forecast']=result.predict(start=90,end=103,dynamic=True)
```

```python
df[['Sales','forecast']].plot(figsize=(15,6))
```

```python
model1=sm.tsa.statespace.SARIMAX(df['Sales'],order=(1,1,1),seasonal_order=(1,1,1,12))
result1=model1.fit()
```
